```python
import pandas as pd
import numpy as np
import os
from datetime import datetime

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 120)

RAW_PATH = "../data/raw/"

RAW_PATH
```

```python
raw_files = os.listdir(RAW_PATH)
raw_files
```

```python
def leer_raw(nombre_fichero, sep=",", encoding="utf-8"):
    ruta = os.path.join(RAW_PATH, nombre_fichero)
    try:
        df = pd.read_csv(ruta, sep=sep, encoding=encoding)
        print(f"‚úÖ Cargado: {nombre_fichero}")
        print(f"   Filas: {df.shape[0]} | Columnas: {df.shape[1]}\n")
        return df
    except Exception as e:
        print(f"‚ùå Error leyendo {nombre_fichero}: {e}")
        return None
```

```python
def rango_fechas(df, col, nombre=""):
    if df is None:
        print(f"{nombre}: ‚ùå no cargado\n")
        return
    fechas = pd.to_datetime(df[col], errors="coerce")
    print(f"üìÖ {nombre}:")
    print("   M√≠nimo:", fechas.min())
    print("   M√°ximo:", fechas.max(), "\n")
```

```python
rango_fechas(downloads_steam, "Date", "Steam Downloads")
rango_fechas(steam_userstats, "DateReported", "Steam User Stats")
rango_fechas(steam_sales, "Date", "Steam Sales")
rango_fechas(steam_hourly, "hour_reported" if "hour_reported" in steam_hourly.columns else steam_hourly.columns[0], "Steam Concurrents")
rango_fechas(canalB_raw, "Date", "Canal B TOTAL")
rango_fechas(playfab_raw, "Date", "PlayFab")
```
